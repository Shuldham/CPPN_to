```python
import numpy as np


"""Implements fully connected neural network"""
"""Trained with backpropogation"""
class BackPropNetwork:
    
    #instance variables here
    layerCount = 0
    size = None
    weights = []
    def __init__(self,size):
        self.layerCount = len(size) - 1
        self.size = size
        
        #used to hold output info from last iteration
        self.output_data = []
        self.input_data = []
        
        #index varibales used to traverse size
        l1 = 0
        l2 = 1
        
        # weights will now contain an array of matrices where each element 
        # is the weight matrix for each increasing layer
        while(l2 < len(self.size)):
            self.weights.append(numpy.random.normal(scale = .1 , size = (self.size[l2], self.size[l1] + 1)))
            l1 += l1
            l2 += l2
    
    #Transfer/activation function
    def sigmoid(self, x, derivative=False):
        if not derivative:
            return 1/(1+np.exp(-x))
        else:
            out = self.sigmoid(x,False)
            return out*(1-out)
    
    #runs network based on given input data
    #NOTE: can use numpy vstack to create a row of ones on bottom of matrix, makes bias easy to implement
    #input is passed as a numpy array
    def run(self, inputs):
        
        lnCases = inputs.shape[0] #returns number of rows in input
        
        #clear out all previous values lists
        self.output_data = []
        self.input_data = []
        
        #now, run the network
        #cases to consider: if its input, you pull data from inputs
        #if its hidden, you pull data from the previous layer
        for index in range(self.layerCount):
            if index == 0:
                #input is transposed and then stacked on top of a row on 1s (bias)
                #1 is # of rows, lnCases is # of cols, this would be lnCases because
                #number of cols is equal to what was before the number of rows
                layerInput = self.weights[0].dot(np.vstack([inputs.T, np.ones(1,lnCases)]))
            else:
                #does same thing but uses last layer's output instead of inputs
                layerInput = self.weights[index].dot(np.vstack([self.output_data[-1].T, np.ones(1,lnCases)]))
            
            #append input (before activation) and output data to running list
            self.input_data.append(layerInput)
            self.output_data.append(self.sigmoid(layerInput))
        
        return self.output_data[-1].T
            

        
```
